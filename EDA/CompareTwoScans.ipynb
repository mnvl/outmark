```python
import numpy as np
import matplotlib.pyplot as plt
import nibabel

plt.style.use('seaborn-deep')
```

```python
data_dir = "/home/mel/Downloads/"

image1 = nibabel.load(data_dir + "img0001.nii.gz")
image2 = nibabel.load(data_dir + "img0002.nii.gz")

label1 = nibabel.load(data_dir + "label0001.nii.gz")
label2 = nibabel.load(data_dir + "label0002.nii.gz")
```

# Распределение интенсивностей

Синий -- первая картинка, зеленый -- вторая.

```python
plt.hist([image1.get_data().flatten(), image2.get_data().flatten()])
plt.show()
```

```python
plt.hist([image1.get_data().flatten(), image2.get_data().flatten()], range = [-500, 500])
plt.show()
```

```python
plt.hist([image1.get_data().flatten(), image2.get_data().flatten()], range = [500, 1000])
plt.show()
```

```python
plt.hist([image1.get_data().flatten(), image2.get_data().flatten()], range = [-1000, -500])
plt.show()
```

# Распределение интенсивностей по классам

```python
classnames = [
    "(1) spleen",
    "(2) right kidney",
    "(3) left kidney",
    "(4) gallbladder",
    "(5) esophagus",
    "(6) liver",
    "(7) stomach",
    "(8) aorta",
    "(9) inferior vena cava",
    "(10) portal vein and splenic vein",
    "(11) pancreas",
    "(12) right adrenal gland",
    "(13) left adrenal gland"]

plt.figure(figsize=(8, 100))
for i, n in enumerate(classnames):
    plt.subplot(len(classnames), 1, i + 1)
    plt.hist([image1.get_data().flatten()[label1.get_data().flatten() == i + 1],
          image2.get_data().flatten()[label2.get_data().flatten() == i + 1]])
    plt.title(n)
plt.show()
```
